```python
import pandas as pd


raw_data = pd.read_csv("ext_list_September_2024.csv", sep=";")
columns = raw_data.columns[2:]
columns = [column[5:] for column in columns]
new_columns = raw_data.columns[:2].to_list() + columns
raw_data.columns = new_columns

raw_data["subject_areas"] = None
columns = raw_data.columns

for index, row in raw_data.iterrows():
    subject_areas = []
    for i_col, column in enumerate(columns):
        if i_col > 1 and not pd.isna(row[column]):
            subject_areas.append(column)

    raw_data.at[index, "subject_areas"] = "; ".join(subject_areas)

raw_data = raw_data[["ISSN", "EISSN", "subject_areas"]]
raw_data.columns = ["issn", "eissn", "subject_areas"]

raw_data.to_csv("subject_areas.csv", index=False)

raw_data.head(70).tail(10)
```

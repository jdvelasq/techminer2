```python
# Create a dataframe of 20 rows with five columns labeled as the 
# first upper case letters of the alphabet. The dataframe is 
# populated with zeros as integer values.


import pandas as pd  # type: ignore
import numpy as np

df = pd.DataFrame(
    np.zeros((20, 5)),
    columns=[chr(i) for i in range(ord("A"), ord("A") + 5)],
    dtype=int,
)

# For each row in the dataframe, assign a value of 1, at random,
# to one or more columns.

for i in range(len(df)):
    for k in range(0, np.random.randint(1, 5)):
        df.iloc[i, np.random.randint(0, 5)] = 1

df['B'] = 1 - df['A']
# df.loc[4, 'B'] = 1 - df.loc[4, 'B']
# df.loc[8, 'B'] = 1 - df.loc[8, 'B']`
```

```python
# assigns to the varieble 'corr', the auto-correlation bewteen the columns of the dataframe df

corr = df.corr()
corr.round(3)
```

```python
# compute the similarity between the columns of the dataframe df

sim = 1 - np.sqrt(0.5 * (1 - corr))
sim
```

```python
# compute the correlation between the columns of the dataframe df
# where the dataframe is binary

corr_bin = df.corr(method='spearman')
corr_bin.round(3)

# compute the similarity between the columns of the dataframe df
# where the dataframe is binary

sim_bin = 1 - np.sqrt(0.5 * (1 - corr_bin))


```
